```python
dept=pd.read_csv('courses_data.csv')['개설학과'].unique()
dept
```

```python
df=df[df['개설학과'].isin(dept)]
df
```

```python
df.loc[df['과정'] == '전공심화', '개설\n학년'] = 4
df
```

```python
df.dropna(subset=['건물명'],inplace=True)
df
```

```python
df.loc[df['개설학과']!='빅데이터과','개설학과']='소프트웨어융합과'
df
```

```python
df=df[[ '개설학과','교과목명', '개설\n학년', '수강\n분반','수강\n인원', '제한\n인원', '강좌\n대표교수','교과목\n학점']]
df
```

```python
df.loc[df['개설\n학년'] == 3, '교과목\n학점'] = df['교과목\n학점'] * 2
df
```

```python
df
```

```python
df.columns=list(map(lambda x : x.replace("\n",""),df.columns))
df
```

```python
for i in df.select_dtypes('number'):
    df[i]=df[i].astype(int)
df
```

```python
df['수강분반']=df['수강분반'].apply(lambda x : x[1])
df
```
